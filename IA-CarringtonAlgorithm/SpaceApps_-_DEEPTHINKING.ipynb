### Prediccion de velocidad de viento solar

```python
#The input and output data of the neural network are selected
X_velocidad = sample.iloc[:,0:8]
Y_velocidad = sample.iloc[:,9:10]
```

```python
#It is verified that they have the same sizes
X_velocidad.shape, Y_velocidad.shape 

#The network model is created
model = Sequential()
#It has a layer of 8 neurons and a relu activation function.
model.add(Dense(8, activation='relu'))
#A layer of 1 neuron with the name dense_2 is added
model.add(Dense(units=1, name = 'dense_2'))
#The RMSprop optimizer is used with a learning ratio of 0.08 and a Gradient Value Clipping of 1
optmizador = tf.keras.optimizers.RMSprop(lr=0.008, clipvalue=1.0)
#the model is compiled with a cost function mse where it should show us the following metrics: 'mae' and 'mse'
model.compile(loss='mse', optimizer=optmizador, metrics=['mae', 'mse'])
#We flatten the layers
model.add(Flatten())
```

```python
#Entrenamiento del modelo
model.fit(X_velocidad, Y_velocidad, epochs=10000, batch_size=10)
```

```python
# Realizar predicciones
predictions_velocity = model.predict(X_velocidad)
print(predictions_velocity.transpose())
```
